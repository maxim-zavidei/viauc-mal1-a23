#### 1. Data Processing

```python
# Loading the meso dataset:
import pandas as pd
meso_raw_df = pd.read_csv('data/Risoe/meso_Risoe.csv')

pd.set_option('display.max_columns', None)
print(meso_raw_df.shape, f"- {meso_raw_df.shape[0]} rows with {meso_raw_df.shape[1]} features")
meso_raw_df.head()
```

```python
# Loading the mast dataset:
import xarray as xr

ds = xr.open_dataset('data/Risoe/risoe_m_all.nc')
mast_raw_df = ds.to_dataframe()

# Reset the index for a clean DataFrame
mast_raw_df = mast_raw_df.reset_index()

pd.set_option('display.max_columns', None)
print(mast_raw_df.shape, f"- {mast_raw_df.shape[0]} rows with {mast_raw_df.shape[1]} features")
mast_raw_df.head()
```

#### Explanation:
In point (b) we have to decide which wind features we will continue to work with. The columns from which to select are "ws44", "ws44_qc", "ws77", "ws77_qc", "wd77", "wd77_qc", "ws125", "ws125_qc", "wd125" and "wd125_qc".

The prefix ws stands for wind speed and the prefix for wd stands for wind direction.
The suffix number in each column represents the altitude at which the observation was taken.

As point (b) states, our most important goal is to chose an altitude which will have an amount of observations that properly cover of all seasons of the year.

To achieve this we will divide the dataset into 3 smaller ones base on the altitude, and perform some analysis on them to decide which one to use.

```python
# Create a new dataframe with only the wind features measured at altitude 44
mast_clean_wind_at_44_df = mast_raw_df[['time', 'ws44', 'ws44_qc']].copy().dropna()

# Create a new dataframe with only the wind features measured at altitude 77
mast_clean_wind_at_77_df = mast_raw_df[['time', 'ws77', 'ws77_qc', 'wd77', 'wd77_qc']].copy().dropna()

# Create a new dataframe with only the wind features measured at altitude 125
mast_clean_wind_at_125_df = mast_raw_df[['time', 'ws125', 'ws125_qc', 'wd125', 'wd125_qc']].copy().dropna()
```

```python
import matplotlib.pyplot as plt

# Step 1: Parse the 'time' column as datetime
mast_clean_wind_at_44_df['time'] = pd.to_datetime(mast_clean_wind_at_44_df['time'])
mast_clean_wind_at_77_df['time'] = pd.to_datetime(mast_clean_wind_at_77_df['time'])
mast_clean_wind_at_125_df['time'] = pd.to_datetime(mast_clean_wind_at_125_df['time'])

# Step 2: Create a new column for the season
# Define a function to get the season from a date
def get_season(date):
    month = date.month
    if 3 <= month <= 5:
        return 'spring'
    elif 6 <= month <= 8:
        return 'summer'
    elif 9 <= month <= 11:
        return 'autumn'
    else:
        return 'winter'

mast_clean_wind_at_44_df['season'] = mast_clean_wind_at_44_df['time'].apply(get_season)
mast_clean_wind_at_77_df['season'] = mast_clean_wind_at_77_df['time'].apply(get_season)
mast_clean_wind_at_125_df['season'] = mast_clean_wind_at_125_df['time'].apply(get_season)

# Step 3: Group by season and count the number of observations
season_counts_for_wind_at_44 = mast_clean_wind_at_44_df.groupby('season').size()
season_counts_for_wind_at_77 = mast_clean_wind_at_77_df.groupby('season').size()
season_counts_for_wind_at_125 = mast_clean_wind_at_125_df.groupby('season').size()

# Step 4: Create the bar charts

# Create the figure and three subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Create the bar plots for each subplot
ax1.bar(season_counts_for_wind_at_44.index, season_counts_for_wind_at_44.values)
ax1.set_title('Observations by season for wind at altitude 44')

ax2.bar(season_counts_for_wind_at_77.index, season_counts_for_wind_at_77.values)
ax2.set_title('Observations by season for wind at altitude 77')

ax3.bar(season_counts_for_wind_at_125.index, season_counts_for_wind_at_125.values)
ax3.set_title('Observations by season for wind at altitude 125')

# Add annotations
for i, v in enumerate(season_counts_for_wind_at_44.values):
    ax1.text(i, v, str(v), ha='center')

for i, v in enumerate(season_counts_for_wind_at_77.values):
    ax2.text(i, v, str(v), ha='center')

for i, v in enumerate(season_counts_for_wind_at_125.values):
    ax3.text(i, v, str(v), ha='center')

# Show the plot
plt.show()
```

#### Explanation:
As we can see the best category which has the most evenly distributed number of observations by season and the most amount of observation in each season is wind speed 77.
Therefore in our existing mast dataframe we can drop the features related to the wind speeds of 44 and 125.

```python
mast_clean_df = mast_raw_df.drop(columns = ['ws44', 'ws44_qc', 'ws125', 'ws125_qc', 'wd125', 'wd125_qc']).copy()

# Drop any rows containing null values in columns related to wind speed 77
mast_clean_df = mast_clean_df.dropna(subset = ['ws77', 'ws77_qc', 'wd77', 'wd77_qc'])

print(mast_clean_df.shape, f"- {mast_clean_df.shape[0]} rows with {mast_clean_df.shape[1]} features")
mast_clean_df.head()
```

##### a) Cleanse wind speeds and wind directions

```python
# Remove any rows where wind speeds are negative
mast_clean_df = mast_clean_df[mast_clean_df['ws77'] >= 0]

# Remove any rows where wind direction is not in interval [0, 360[
mast_clean_df = mast_clean_df[mast_clean_df['wd77'] >= 0 & (mast_clean_df['wd77'] < 360)]
print(mast_clean_df.shape, f"- {mast_clean_df.shape[0]} rows with {mast_clean_df.shape[1]} features")
```

##### b) Selected ws77 and wd77 based on the previous points

##### c) Find the meso-signals closest in height to the mast-data-signal

#### Explanation:
In the meso dataset we have observations of wind features at the following heights: 60, 80, 100, 120, 140, 160, 180, 200, 220 (m)

Our selected mast height is 77, therefore the closest height in the meso dataset is 80.

```python
# Keep only the columns which hold values at height 80
meso_clean_df = meso_raw_df[['TIMESTAMP', 'HGT', 'RHO080', 'TC080', 'WSP080', 'WDIR080', 'PRS080', 'TK080', 'TKE_PBL080', 'RH080']].copy()

# Drop any rows which contain null values
meso_clean_df = meso_clean_df.dropna()

print(meso_clean_df.shape, f"- {meso_clean_df.shape[0]} rows with {meso_clean_df.shape[1]} features")
meso_clean_df.head()
```

##### d) Convert the mast data from DK time to UTC time

```python
meso_clean_df['TIMESTAMP'] = meso_clean_df['TIMESTAMP'].apply(lambda x: pd.to_datetime(x)).dt.tz_localize('UTC')

mast_clean_df['time'] = mast_clean_df['time'].dt.tz_localize('UTC')
mast_clean_df['time'] = mast_clean_df['time'].dt.tz_convert('Europe/Copenhagen')
mast_clean_df['time'] = mast_clean_df['time'].dt.tz_convert('UTC')

print(mast_clean_df.shape, f"- {mast_clean_df.shape[0]} rows with {mast_clean_df.shape[1]} features")
mast_clean_df.tail()
```

e) Resample the mast dataset to have the same frequency as the meso data

```python
mast_resampled_df = mast_clean_df.resample('60min', on='time').mean()
mast_resampled_df.head()
```

f) Find the overlapping timestamps between the meso data and the resampled mast data

```python
meso_clean_df = meso_clean_df.set_index('TIMESTAMP')
result = pd.concat([meso_clean_df, mast_resampled_df], axis=1, join='inner')
result
```
