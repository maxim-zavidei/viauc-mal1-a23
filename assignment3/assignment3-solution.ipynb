#### 1. Data Processing

```python
# Loading the meso dataset:
import pandas as pd
meso_raw_df = pd.read_csv('data/Risoe/meso_Risoe.csv')

pd.set_option('display.max_columns', None)
print(meso_raw_df.shape, f"- {meso_raw_df.shape[0]} rows with {meso_raw_df.shape[1]} features")
meso_raw_df.head()
```

```python
# Loading the mast dataset:
import xarray as xr

ds = xr.open_dataset('data/Risoe/risoe_m_all.nc')
mast_raw_df = ds.to_dataframe()

# Reset the index for a clean DataFrame
mast_raw_df = mast_raw_df.reset_index()

pd.set_option('display.max_columns', None)
print(mast_raw_df.shape, f"- {mast_raw_df.shape[0]} rows with {mast_raw_df.shape[1]} features")
mast_raw_df.head()
```

#### Explanation:
In point (b) we have to decide which wind features we will continue to work with. The columns from which to select are "ws44", "ws44_qc", "ws77", "ws77_qc", "wd77", "wd77_qc", "ws125", "ws125_qc", "wd125" and "wd125_qc".

The prefix ws stands for wind speed and the prefix for wd stands for wind direction.
The suffix number in each column represents the altitude at which the observation was taken.

As point (b) states, our most important goal is to chose an altitude which will have an amount of observations that properly cover of all seasons of the year.

To achieve this we will divide the dataset into 3 smaller ones base on the altitude, and perform some analysis on them to decide which one to use.

```python
# Create a new dataframe with only the wind features measured at altitude 44
mast_clean_wind_at_44_df = mast_raw_df[['time', 'ws44', 'ws44_qc']].copy().dropna()

# Create a new dataframe with only the wind features measured at altitude 77
mast_clean_wind_at_77_df = mast_raw_df[['time', 'ws77', 'ws77_qc', 'wd77', 'wd77_qc']].copy().dropna()

# Create a new dataframe with only the wind features measured at altitude 125
mast_clean_wind_at_125_df = mast_raw_df[['time', 'ws125', 'ws125_qc', 'wd125', 'wd125_qc']].copy().dropna()
```

```python
import matplotlib.pyplot as plt

# Step 1: Parse the 'time' column as datetime
mast_clean_wind_at_44_df['time'] = pd.to_datetime(mast_clean_wind_at_44_df['time'])
mast_clean_wind_at_77_df['time'] = pd.to_datetime(mast_clean_wind_at_77_df['time'])
mast_clean_wind_at_125_df['time'] = pd.to_datetime(mast_clean_wind_at_125_df['time'])

# Step 2: Create a new column for the season
# Define a function to get the season from a date
def get_season(date):
    month = date.month
    if 3 <= month <= 5:
        return 'spring'
    elif 6 <= month <= 8:
        return 'summer'
    elif 9 <= month <= 11:
        return 'autumn'
    else:
        return 'winter'

mast_clean_wind_at_44_df['season'] = mast_clean_wind_at_44_df['time'].apply(get_season)
mast_clean_wind_at_77_df['season'] = mast_clean_wind_at_77_df['time'].apply(get_season)
mast_clean_wind_at_125_df['season'] = mast_clean_wind_at_125_df['time'].apply(get_season)

# Step 3: Group by season and count the number of observations
season_counts_for_wind_at_44 = mast_clean_wind_at_44_df.groupby('season').size()
season_counts_for_wind_at_77 = mast_clean_wind_at_77_df.groupby('season').size()
season_counts_for_wind_at_125 = mast_clean_wind_at_125_df.groupby('season').size()

# Step 4: Create the bar charts

# Create the figure and three subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Create the bar plots for each subplot
ax1.bar(season_counts_for_wind_at_44.index, season_counts_for_wind_at_44.values)
ax1.set_title('Observations by season for wind at altitude 44')

ax2.bar(season_counts_for_wind_at_77.index, season_counts_for_wind_at_77.values)
ax2.set_title('Observations by season for wind at altitude 77')

ax3.bar(season_counts_for_wind_at_125.index, season_counts_for_wind_at_125.values)
ax3.set_title('Observations by season for wind at altitude 125')

# Add annotations
for i, v in enumerate(season_counts_for_wind_at_44.values):
    ax1.text(i, v, str(v), ha='center')

for i, v in enumerate(season_counts_for_wind_at_77.values):
    ax2.text(i, v, str(v), ha='center')

for i, v in enumerate(season_counts_for_wind_at_125.values):
    ax3.text(i, v, str(v), ha='center')

# Show the plot
plt.show()
```

#### Explanation:
As we can see the best category which has the most evenly distributed number of observations by season and the most amount of observation in each season is wind speed 77.
Therefore in our existing mast dataframe we can drop the features related to the wind speeds of 44 and 125.

```python
mast_clean_df = mast_raw_df.drop(columns = ['ws44', 'ws44_qc', 'ws125', 'ws125_qc', 'wd125', 'wd125_qc']).copy()

# Drop any rows containing null values in columns related to wind speed 77
mast_clean_df = mast_clean_df.dropna(subset = ['ws77', 'ws77_qc', 'wd77', 'wd77_qc'])

print(mast_clean_df.shape, f"- {mast_clean_df.shape[0]} rows with {mast_clean_df.shape[1]} features")
mast_clean_df.head()
```

##### a) Cleanse wind speeds and wind directions

```python
# Remove any rows where wind speeds are negative
mast_clean_df = mast_clean_df[mast_clean_df['ws77'] >= 0]

# Remove any rows where wind direction is not in interval [0, 360[
mast_clean_df = mast_clean_df[mast_clean_df['wd77'] >= 0 & (mast_clean_df['wd77'] < 360)]
print(mast_clean_df.shape, f"- {mast_clean_df.shape[0]} rows with {mast_clean_df.shape[1]} features")
```

##### b) Selected ws77 and wd77 based on the previous points

##### c) Find the meso-signals closest in height to the mast-data-signal

#### Explanation:
In the meso dataset we have observations of wind features at the following heights: 60, 80, 100, 120, 140, 160, 180, 200, 220 (m)

Our selected mast height is 77, therefore the closest height in the meso dataset is 80.

```python
# Keep only the columns which hold values at height 80
meso_clean_df = meso_raw_df[['TIMESTAMP', 'HGT', 'RHO080', 'TC080', 'WSP080', 'WDIR080', 'PRS080', 'TK080', 'TKE_PBL080', 'RH080']].copy()

# Drop any rows which contain null values
meso_clean_df = meso_clean_df.dropna()

print(meso_clean_df.shape, f"- {meso_clean_df.shape[0]} rows with {meso_clean_df.shape[1]} features")
meso_clean_df.head()
```

##### d) Convert the mast data from DK time to UTC time

```python
meso_clean_df['TIMESTAMP'] = meso_clean_df['TIMESTAMP'].apply(lambda x: pd.to_datetime(x)).dt.tz_localize('UTC')

mast_clean_df['time'] = mast_clean_df['time'].dt.tz_localize('UTC')
mast_clean_df['time'] = mast_clean_df['time'].dt.tz_convert('Europe/Copenhagen')
mast_clean_df['time'] = mast_clean_df['time'].dt.tz_convert('UTC')

print(mast_clean_df.shape, f"- {mast_clean_df.shape[0]} rows with {mast_clean_df.shape[1]} features")
mast_clean_df.tail()
```

##### e) Resample the mast dataset to have the same frequency as the meso data

```python
mast_resampled_df = mast_clean_df.resample('60min', on='time').mean()
mast_resampled_df.head()
```

##### f) Find the overlapping timestamps between the meso data and the resampled mast data

```python
meso_clean_df = meso_clean_df.set_index('TIMESTAMP')
overlapped_meso_mast_df = pd.concat([meso_clean_df, mast_resampled_df], axis=1, join='inner')
overlapped_meso_mast_df
```

#### 2. Data Analysis

##### a) Fit the different wind speeds into a Weibull distribution

```python
import numpy as np
from scipy.stats import weibull_min
import matplotlib.pyplot as plt

# Extract the column to be fit into a Weibull distribution
wind_speed_mast_clean = mast_clean_df['ws77'].dropna()
wind_speed_meso_clean = meso_clean_df['WSP080'].dropna()
wind_speed_mast_resampled = mast_resampled_df['ws77'].dropna()

# Fit the data to a Weibull distribution
shape_mast_clean, loc_mast_clean, scale_mast_clean = weibull_min.fit(wind_speed_mast_clean, loc=0)
shape_meso_clean, loc_meso_clean, scale_meso_clean = weibull_min.fit(wind_speed_meso_clean, loc=0)
shape_mast_resampled, loc_mast_resampled, scale_mast_resampled = weibull_min.fit(wind_speed_mast_resampled, loc=0)

# Generate Weibull distribution using fitted parameters
weibull_dist_mast_clean = weibull_min(shape_mast_clean, loc=loc_mast_clean, scale=scale_mast_clean)
weibull_dist_meso_clean = weibull_min(shape_meso_clean, loc=loc_meso_clean, scale=scale_meso_clean)
weibull_dist_mast_resampled = weibull_min(shape_mast_resampled, loc=loc_mast_resampled, scale=scale_mast_resampled)

# Generate the probability density function values
x_mast_clean = np.linspace(min(wind_speed_mast_clean), max(wind_speed_mast_clean), 100)
pdf_mast_clean = weibull_dist_mast_clean.pdf(x_mast_clean)

x_meso_clean = np.linspace(min(wind_speed_meso_clean), max(wind_speed_meso_clean), 100)
pdf_meso_clean = weibull_dist_meso_clean.pdf(x_meso_clean)

x_mast_resampled = np.linspace(min(wind_speed_mast_resampled), max(wind_speed_mast_resampled), 100)
pdf_mast_resampled = weibull_dist_mast_resampled.pdf(x_mast_resampled)

# Plot the original data and the fitted Weibull distribution
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.hist(wind_speed_mast_clean, bins=20, density=True, alpha=0.6, color='g', label='Data')
plt.plot(x_mast_clean, pdf_mast_clean, 'r-', lw=2, label='Fitted Weibull (mast_clean_df)')
plt.legend(loc='best')

plt.subplot(132)
plt.hist(wind_speed_meso_clean, bins=20, density=True, alpha=0.6, color='g', label='Data')
plt.plot(x_meso_clean, pdf_meso_clean, 'r-', lw=2, label='Fitted Weibull (meso_clean_df)')
plt.legend(loc='best')

plt.subplot(133)
plt.hist(wind_speed_mast_resampled, bins=20, density=True, alpha=0.6, color='g', label='Data')
plt.plot(x_mast_resampled, pdf_mast_resampled, 'r-', lw=2, label='Fitted Weibull (mast_resampled_df)')
plt.legend(loc='best')

plt.tight_layout()
plt.show()
```

Above is presented the correlation table for the mast data set. The first significant observation we can notice is that the temperature measurements at

All the values are 98% or greater. The interpretation of these numbers signifies that for a substantial number of meteorological measurements whenever

#### 3. Model Development

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline 

X = overlapped_meso_mast_df.drop(columns=['ws77', 'WSP080'])
y = overlapped_meso_mast_df['WSP080']

# Split the dataset into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Initialize a set of alphas
alphas = 10**np.linspace(-10, 10, 100)
```

```python
# Create a pipeline with StandardScaler and RidgeCV
ridgecv = make_pipeline(StandardScaler(), RidgeCV(alphas=alphas, scoring='neg_mean_squared_error'))
ridgecv.fit(X_train, y_train)

# Access the chosen alpha value from RidgeCV
chosen_alpha = ridgecv.named_steps['ridgecv'].alpha_
print(chosen_alpha)
```

```python
# Create a Ridge model using the chosen alpha from RidgeCV
ridge_model = Ridge(alpha=ridgecv.named_steps['ridgecv'].alpha_)
ridge_model.fit(X_train, y_train)

# Calculate the test MSE using the trained Ridge model
mse = mean_squared_error(y_test, ridge_model.predict(X_test))
print(mse)
```

```python
ridge_model.fit(X, y)
pd.Series(ridge_model.coef_, index = X.columns)
```

```python
# Print the R^2
print("R^2 on train data is {} and on test data is {}".format(ridgecv.score(X_train, y_train), ridgecv.score(X_test,y_test)))
```

```python
import matplotlib.pyplot as plt

# Make predictions using the testing set
y_pred = ridge_model.predict(X_test)

# Plotting Actual vs. Predicted values
plt.figure(figsize=(12, 6))

# Scatter plot of actual and predicted values
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual ws77")
plt.ylabel("Predicted ws77")
plt.title("Actual vs. Predicted ws77")

# Plot a diagonal (identity) line, where the predicted values are identical to the actual values
max_limit = max(y_test.max(), y_pred.max())
min_limit = min(y_test.min(), y_pred.min())  # Choose appropriate limits for x and y axis
plt.plot([min_limit, max_limit], [min_limit, max_limit], 'k--')

plt.grid(True)
plt.show()

# Plotting residuals (difference between actual and predicted values)
plt.figure(figsize=(12, 6))

residuals = y_test - y_pred
plt.scatter(y_test, residuals, alpha=0.5)
plt.hlines(y=0, xmin=y_test.min(), xmax=y_test.max(), colors='k', linestyles='dashed', lw=2)
plt.xlabel("Actual ws77")
plt.ylabel("Residuals")
plt.title("Residuals of the Predictions (Actual - Predicted)")
plt.grid(True)
plt.show()

# Histogram of residuals
plt.figure(figsize=(12, 6))
plt.hist(residuals, bins=20, alpha=0.5, edgecolor='k')
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Distribution of Residuals")
plt.grid(True)
plt.show()
```
